#**Finding where do the guests come from ?**

```python
#Creating a new dataframe for guest country

guest_country= update_hotel_df[update_hotel_df['is_canceled']==0]['country'].value_counts().reset_index()
guest_country.columns=['country','Number of guests']
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import plotly.express as px
```

```python
guest_country
```

```python
total_guests= guest_country['Number of guests'].sum()
guest_country['guest in %'] = round(guest_country['Number of guests']/ total_guests * 100,2)
```

```python
guest_country
```

```python
#Ploting a map to find the highest guest are coming from which countries

guest_map = px.choropleth(guest_country,
                          locations=guest_country['country'],
                          color=guest_country['Number of guests'],
                          hover_name=guest_country['country'],
                          title='home country of guests')

guest_map.show()
```

###This graph shows that from which country most of the guest are coming. Almost 50% of guests are coming from Portugal,Great Britain,France.

#**Miss interpreting data**

```python
resort= update_hotel_df[(update_hotel_df['hotel']=='Resort Hotel') & (update_hotel_df['is_canceled']== 0)]
city= update_hotel_df[(update_hotel_df['hotel']=='City Hotel') & (update_hotel_df['is_canceled'] == 0)]
```

```python
resort
```

```python
resort_hotel= resort.groupby(['arrival_date_month'])['adr'].mean().reset_index()
resort_hotel
```

```python
city_hotel=city.groupby(['arrival_date_month'])['adr'].mean().reset_index()
city_hotel
```

```python
final_df=resort_hotel.merge(city_hotel, on='arrival_date_month')
final_df.columns=['month','price_for_resort','price_for_city_hotel']
final_df
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Convert the 'month' column to categorical with custom ordering
final_df['month'] = pd.Categorical(final_df['month'], categories=month_order, ordered=True)

# Sort the DataFrame by the 'month' column
final_df = final_df.sort_values('month')
```

```python
final_df
```

```python
plt.figure(figsize=(5,3))
px.line(final_df, x='month',
        y=['price_for_resort', 'price_for_city_hotel'],
        title='room price per night over a month')
```

###In this graph we can see the price of resort increasing after the ending of april month. And in august the resort price was the highest. After that the price of resort price were start decreasing

###There was not much change in the price of city hotel.

###**How much do guests pays for a room per night ?**

```python
update_hotel_df['adr_updated']= update_hotel_df['adr']/(update_hotel_df['adults']+update_hotel_df['children'])
```

```python
valid_guest = update_hotel_df.loc[update_hotel_df['is_canceled']==0]

prices = valid_guest[['hotel','reserved_room_type','adr_updated']].sort_values('reserved_room_type')
prices
```

```python
#Plotting a boxplot to find outliers

plt.figure(figsize=(11,7))
sns.boxplot(x='reserved_room_type',
            y='adr_updated',
            hue='hotel',
            data=prices)
plt.title('Price of room type per night and person')
plt.xlabel('Room type')
plt.ylabel('Price [EUR]')

plt.ylim(0,160)
plt.show()
```

###Those values which are more than 143 is all outliers inside the city hotel.

###Those values which are more than 116 is all outliers inside the resort hotel.

```python
prices_c= prices[prices['reserved_room_type']=='C']
prices_c
```

```python
#Filtering the data for both City hotels and Resort hotels

prices_city=prices_c[prices_c['hotel']=='City Hotel']
prices_resort=prices_c[prices_c['hotel']=='Resort Hotel']
```
